```python
import sympy as spy
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y=spy.symbols('x y')
fun=8*x**2+1*y**2


gradfun=[spy.diff(fun,var) for var in (x,y)]


numgradfun=spy.lambdify([x,y],gradfun)
```

```python
X,Y=np.meshgrid(np.arange(-10,11),np.arange(-10,11))
graddat=numgradfun(X,Y)

plt.figure(figsize=(14, 12), dpi=80)
plt.quiver(X,Y,graddat[0],graddat[1])
plt.show()
```
