```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
        return x**2+2*x+2  # xÂ² + 2x + 2
```

```python
x = np.linspace(-6, 6)
```

```python
plt.plot(x, f(x), color='purple', label='Function')
```

```python
h = 0.01
c1 = 0
cf = c1+h
```

```python
y1 = f(c1)
y2 = f(cf)
```

```python
derivada = (y2 - y1)/h
```

```python
print(f'A derivada aproximada Ã© {derivada}')
```

```python
def dev(x):
          return 2*x+2
```

```python
plt.plot(x, f(x), color='purple', label='Function')
plt.plot(x, dev(x), color='red', label='Dev')
```
