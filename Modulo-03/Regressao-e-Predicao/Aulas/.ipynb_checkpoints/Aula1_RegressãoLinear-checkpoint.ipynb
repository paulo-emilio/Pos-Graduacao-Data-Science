# ***REGRESSÃO LINEAR - PARTE 1 (Definindo Regressão Linear)*** 


A análise de regressão linear é usada em várias áreas, como ciência da computação, finanças, economia, engenharia, meteorologia e biologia. É usado para prever o comportamento de um sistema com base em dados coletados anteriores.

A regressão linear é frequentemente usada para prever valores futuros. Por exemplo, os economistas usam a regressão linear para prever a taxa de inflação, os vendedores para prever as vendas futuras e os analistas financeiros para avaliar o retorno esperado de um investimento.

Modelos de regressão linear também são úteis para explicar o quão e por que variáveis ​​diferentes estão relacionadas. Por exemplo, se você quiser prever o preço de um carro com base em diferentes variáveis ​​de entrada, como marca, ano e quilometragem, é possível usar uma regressão linear para explicar como cada variável contribui para o preço final.

Em resumo, modelos de regressão linear podem ser úteis para prever o futuro e explicar o comportamento de variáveis ​​dependentes. Eles são simples de entender e podem treinar rapidamente para produzir resultados precisos.

# **REGRESSÃO LINEAR - PARTE 2 (Estimativas com regressão linear)**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# variável preditora
X = np.array([ 220, 220, 220, 220, 220, 225, 225, 225, 225, 225, 230, 230, 230, 230, 230, 235, 235, 235, 235, 235 ])
# variável alvo
y = np.array([ 137, 137, 137, 136, 135, 135, 133, 132, 133, 133, 128, 124, 126, 129, 126, 122, 122, 122, 119, 122 ])
```

```python
# é necessário adicionar uma constante a matriz X
X_sm = sm.add_constant(X)
```

```python
# OLS vem de Ordinary Least Squares e o método fit irá treinar o modelo
results = sm.OLS(y, X_sm).fit()
```

```python
# mostrando as estatísticas do modelo
print(results.summary())
```

# **REGRESSÃO LINEAR - PARTE 3 (Análise de Predição com Regressão Linear)**

Contrato de trabalho com uma empresa de comércio eletrônico com sede na cidade de Recife.

Investir em aplicativos móveis ou em e-commerce?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
clientes = pd.read_csv("../dados/Ecommerce Customers.csv")
clientes.head()
```

```python
# sns.pairplot(clientes)
pd.plotting.scatter_matrix(clientes, figsize=(10, 10))
plt.show()
```

```python
sns.lmplot(x='Length of Membership', y='Yearly Amount Spent', data=clientes)
```

```python
y = clientes['Yearly Amount Spent']
```

```python
X= clientes[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.3,random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
prediction = lm.predict(X_test)
```

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

# **REGRESSÃO LINEAR - PARTE 4 (Inferência Casual)**


A inferência causal é a ciência de inferir a presença e a magnitude
das relações de causa e efeito a partir dos dados.

Como estatísticos, epidemiologistas, sociólogos, etc., e de fato como
seres humanos, é algo sobre o qual sabemos muito.

CI consiste numa família de métodos que buscam respondem o «porquê» das coisas acontecerem. Os métodos comuns como análise de regressão se preocupam em quantificar correlações, como mudanças em X estão associadas em mudanças em Y. Já os métodos de CI buscam determinar SE mudanças em X causam mudanças em Y. As abordagens de CI buscam responder perguntas do tipo «por quê» Y muda.

Além disso, o CI pode auxiliar na análise de dados para entender melhor o comportamento dos usuários e qualificar o tráfego que está chegando ao site, ajudando a direcionar o tráfego para páginas mais relevantes. O CI também pode ajudar a identificar padrões de comportamento e uso dos produtos para melhorar a experiência dos usuários.
