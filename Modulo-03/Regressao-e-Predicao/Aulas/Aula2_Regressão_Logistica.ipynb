A regressão não linear é um método de localizar um modelo não linear do relacionamento entre a variável dependente e um conjunto de variáveis independentes. Isso é feito usando algoritmos de estimação iterativos.

Exemplo. A população pode ser prevista com base no tempo? Ao configurar uma equação apropriada, como um modelo logístico de crescimento de população, podemos obter uma boa estimativa do modelo, permitindo fazer predições sobre a população para os momentos que não foram realmente medidos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,2,3,4,5,6,7])
```

```python
y = np.array([0.32,0.12,0.21,0.47,0.52,0.76,0.87,0.96])
```

```python
plt.plot(x,y,'o')
```

```python
""" Ajuste da curva a um polinômio """
# p1 = np.polyfit(x,y,1)  
p2 = np.polyfit(x,y,2)  
p3 = np.polyfit(x,y,3) 
```

```python
""" Cálculo dos coeficientes de determinação dos polinômios de ordem 2 e 3 """
yfit2 = p2[0] * pow(x,2) + p2[1] * x + p2[2] 
yresid2 = y - yfit2 
SQresid = sum(pow(yresid2,2)) 
SQtotal = len(y) * np.var(y) 
R2_2 = 1 - SQresid/SQtotal 
```

```python
yfit3 = p3[0] * pow(x,3) + p3[1] * pow(x,2) + p3[2] * x + p3[3] 
yresid3 = y - yfit3 
SQresid = sum(pow(yresid3,2)) 
SQtotal = len(y) * np.var(y) 
R2_3 = 1 - SQresid/SQtotal 
```

```python
""" Plotagem dos gráficos """
import matplotlib.pyplot as plt
plt.plot(x,y,'o')
plt.plot(x,np.polyval(p2,x),'b--')
plt.plot(x,np.polyval(p3,x),'r:')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# **REGRESSÃO LOGÍSTICA E OUTROS TIPOS - PARTE 3 (Definindo Regressão Logística)**

A regressão logística é uma técnica de análise de dados que usa matemática para encontrar as relações entre dois fatores de dados. Em seguida, essa relação é usada para prever o valor de um desses fatores com base no outro. A previsão geralmente tem um número finito de resultados, como sim ou não.

Por exemplo, digamos que você deseje adivinhar se o visitante do seu site clicará no botão de finalização de compra no carrinho de compras ou não. A análise de regressão logística analisa o comportamento anterior do visitante, como o tempo gasto no site e o número de itens no carrinho. Ela determina que, anteriormente, se os visitantes passassem mais de cinco minutos no site e adicionassem mais de três itens ao carrinho, eles clicavam no botão de finalização de compra. Usando essas informações, a função de regressão logística pode prever o comportamento de um novo visitante do site.

Neste exemplo estaremos trabalhando com um conjunto de dados falso de publicidade, indicando se um usuário de internet específico clicou ou não em uma propaganda. Vamos tentar criar um modelo que preveja se clicará ou não em um anúncio baseado nos recursos desse usuário.

Este conjunto de dados contém os seguintes recursos:

* 'Daily Time Spent on Site': tempo no site em minutos.
* 'Age': idade do consumidor.
* 'Area Income': Média da renda do consumidor na região.
* 'Daily Internet Usage': Média em minutos por di que o consumidor está na internet.
* 'Linha do tópico do anúncio': Título do anúncio.
* 'City': Cidade do consumidor.
* 'Male': Se o consumidor era ou não masculino.
* 'Country': País do consumidor.
* 'Timestamp': hora em que o consumidor clicou no anúncio ou janela fechada.
* 'Clicked on Ad'': 0 ou 1 indicam se clicou ou não no anúncio.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ad_data = pd.read_csv('../dados/advertising.csv')
```

```python
sns.set_style('whitegrid')
ad_data['Age'].hist(bins=30)
plt.xlabel('Age')
```

```python
sns.jointplot(x='Age',y='Area Income',data=ad_data)
```

```python
sns.jointplot(x='Age',y='Daily Time Spent on Site',data=ad_data,color='red',kind='kde');
```

```python
sns.jointplot(x='Daily Time Spent on Site',y='Daily Internet Usage',data=ad_data,color='green')
```

```python
sns.pairplot(ad_data,hue='Clicked on Ad',palette='bwr')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

# **REGRESSÃO LOGÍSTICA E OUTROS TIPOS - PARTE 4 e 5 (Logit e Probit)**


A técnica de regressão logística é uma ferramenta estatística utilizada nas análises preditivas. O interesse em mensurar a probabilidade de um evento ocorrer é extremamente relevante em diversas áreas, como por exemplo em Marketing, Propaganda e Internet, na Aplicação da Lei e Detecção de Fraude, na Assistência Médica, com relação aos Riscos Financeiros e Seguros ou mesmo estudando a Força de Trabalho. É imprescindível elevar o conhecimento sobre quais clientes possuem maior propensão à responder o contato de marketing, quais transações serão fraudulentas, quais e-mails são spam, quem efetivamente fará o pagamento de uma obrigação ou mesmo qual criminoso reincidirá.

Os modelos Logit são uma forma de modelo estatístico usado para prever a probabilidade de ocorrência de um evento. Os modelos Logit também são chamados de modelos de regressão logística. O modelo logit é baseado na função logística (também chamada de função sigmóide), que é usada para modelar situações em que há dois resultados possíveis. A função logística pode ser usada para modelar uma variedade de situações, incluindo variáveis dependentes binárias, variáveis dependentes dicotômicas e dados categóricos.

Os modelos Probit são uma forma de modelo estatístico usado para prever a probabilidade de ocorrência de um evento. Os modelos Probit são semelhantes aos modelos logit, mas são baseados na função probit em vez da função logística. O modelo Probit determina a probabilidade de um item ou evento cair em uma de uma variedade de categorias, estimando a probabilidade de que a observação com características específicas pertença a uma categoria específica. No caso do modelo Probit, a variável dependente é categórica e só pode assumir um dos dois valores, como sim ou não, verdadeiro ou falso.

A seguir estão algumas das principais diferenças entre os modelos Logit e Probit:

*   O modelo logit é usado para modelar as chances de sucesso de um evento como uma função de variáveis independentes, enquanto o modelo probit é usado para determinar a probabilidade de um item ou evento cair em uma de uma variedade de categorias, estimando a probabilidade de que observação com características específicas pertencerá a uma categoria particular.
*   No caso do modelo logit, usamos uma função logística ou sigmoide em vez de Φ, que é uma função de distribuição normal padrão cumulativa.
*   Os modelos de regressão logística também são chamados de modelos logit, enquanto os modelos de regressão probit também são chamados de modelos probit.
*   Os modelos Logit são usados para modelar a distribuição logística, enquanto os modelos probit são usados para modelar a distribuição normal padrão cumulativa.
