## `Confusion Matrix`

<br>

A matriz de confusão é a forma mais simples de visualizar uma performance de um modelo de classificação. Ela indica quantos resultados foram falso positivos, falso negativos, verdadeiro positivos ou verdadeiro negativos.

## `Accuracy`

<br>

A acurácia é a métrica que dá a porcentagem de classificações corretas no dataset. Por exemplo, se temos 1000 observações, onde temos que classificá-las em 2 classes, e o classificador acertou 950 previsões, a acurácia seria de 95%.

## `Precision` & `Recall`


A precisão é uma métrica que dá um foco nos positivos. Ela responde a seguinte pergunta: Das observações que classificamos como positiva, quantas realmente foram positivas?


O recall, também chamado de sensibilidade, dá uma maior ênfase nos negativos. Ele responde a seguinte pergunta: De todos as observações que são positivas, quantas foram classificadas como positivas? Ela é muito relevante quando temos classes com menos observações de treino do que outras.

<br>

![alt_text](http://kflu.github.io/2016/08/26/2016-08-26-visualizing-precision-recall/2016-08-26-visualizing-precision-recall-1.png)

<br><br><br>

## `F1-Score`


O F1-Score é uma média harmônica entre a precisão e o recall. É uma forma de olhar para somente uma métrica para avaliar o modelo. Uma vantagem dessa métrica é que, se a precisão for muito baixa, o F1-score também será baixo. O mesmo vale para o recall.

## `Curva ROC`


Para modelos de classificação que definem uma probabilidade de uma observação pertencer a uma classe, se faz necessário definir um threshold que define que tal observação será daquela classe se obtiver uma probabilidade maior do que o threshold definido. Muitas vezes, determinados como 50%, mas pode ser que outros valores gerem resultados melhores.


A curva ROC (Receiver Operating Characteristic) é utilizada para avalizar a performance do algoritmo de classificação para diferentes threshold's de classificação. Ela é feita utilizando a taxa de falso positivos e a taxa de verdadeiro positivos, para cada valor de threshold.
