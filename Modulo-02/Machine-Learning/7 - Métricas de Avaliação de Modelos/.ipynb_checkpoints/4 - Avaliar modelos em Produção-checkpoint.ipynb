## `Qual métrica avaliar?`

<br>

A melhor métrica que você pode escolher é a métrica que seu negócio está tentando melhorar. Ou seja, o seu modelo de machine learning, antes de tudo, tem um objetivo de negócio, e uma métrica para quantificar se esse objetivo de negócio está sendo alcançado é a métrica ideal para avaliar a performance do seu modelo.

<br><br>



<br><br>

### `Mato alto`

<br>

Um ponto interessante é que, na primeira interação do seu modelo, você está comparando um controle sem machine learning contra um com machine learning. Se o seu modelo se mostra uma opção melhor, a partir daí, essa versão do seu modelo será o controle, e você tentará otimizar ainda mais o seu algoritmo para obter melhores resultados!
