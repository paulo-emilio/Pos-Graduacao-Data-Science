```python
import pandas as pd
```

```python
df = pd.read_csv("http://bit.ly/kaggletrain")
df.head()
```

```python
sex = pd.get_dummies(df['Sex'], dtype=int)
sex.head()
```

```python
#criar embarque dummy
emb = pd.get_dummies(df['Embarked'], dtype=int)
emb.head()
```

```python
#concatenar no df
df = pd.concat([df, emb], axis=1)
df.head()
```

```python
# male vs female só precisa de um para saber tudo (se não for mulher é homem, e vice-versa)
df = pd.get_dummies(df, columns=['Sex'], dtype=int, 
               drop_first=True)  # removendo a coluna Sex e Sex_female
df.head()
```
